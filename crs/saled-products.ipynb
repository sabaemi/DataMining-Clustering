```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from numpy import unique
from numpy import where
import matplotlib.pyplot as plt
%matplotlib inline 
from sklearn.cluster import KMeans
%cd /content/drive/My Drive/Colab Notebooks/
df = pd.read_csv(r'orders.csv')

feature_columns = ['ID_Item','city_name_fa']
df=df[feature_columns]
df1=df["city_name_fa"].unique().tolist()
cut = pd.notnull(df["ID_Item"])
dff=df[cut]
df3=dff["ID_Item"].unique().tolist()
dfcat = df3
df=df.to_dict("records")
x=np.zeros((len(df1),len(dfcat)))
for a in df:
  for i in range(len(df1)):
    if a['city_name_fa']==df1[i]:
      for j in range(len(dfcat)):
        if a['ID_Item']==dfcat[j]:
          x[i][j]=1
distortions = []
K = range(1,9)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(x)
    distortions.append(kmeanModel.inertia_)
plt.figure(figsize=(16,8))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```




KMeans


```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 4, max_iter = 300, random_state = 0)
kmeans.fit(x)
centers = kmeans.cluster_centers_
from matplotlib import pyplot
yhat= kmeans.predict(x)
clusters=unique(yhat)
for i in clusters:
  print("Cluster",i,"countains:")
  for j in range(len(yhat)):
    if yhat[j]==i :
      print("city",df1[j]) 
```

Agglomerative

```python
from sklearn.cluster import AgglomerativeClustering
model= AgglomerativeClustering(n_clusters=4)
yhat=model.fit_predict(x)
clusters=unique(yhat)
for i in clusters:
  print("Cluster",i,"countains:")
  for j in range(len(yhat)):
    if yhat[j]==i :
      print("city",df1[j])
```

```python
from sklearn.cluster import DBSCAN
model= DBSCAN(eps=0.1, min_samples=9)
yhat=model.fit_predict(x)
clusters=unique(yhat)
for i in clusters:
  print("Cluster",i,"countains:")
  for j in range(len(yhat)):
    if yhat[j]==i :
      print("city",df1[j]) 
```
